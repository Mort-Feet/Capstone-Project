# Feature Engineering

```python
# Changing 'Year_Birth' to current age
df['Age'] = 2024 - df['Year_Birth']

# 'Marital Status' has too many values. Convert it to only 'Attached' & 'Single'
df['Relationship_Status'] = df['Marital_Status'].replace({'Married': 'Attached',
                                                          'Together': 'Attached', 
                                                          'Divorced': 'Single',
                                                          'Widow': 'Single',
                                                          'Alone': 'Single',
                                                          'Absurd': 'Single',
                                                          'YOLO': 'Single'})
# Doing the same for 'Education'
df['Education'] = df['Education'].replace({'Graduation': 'Graduate',
                                           'PhD': 'Post-Grad',
                                          'Master': 'Post-Grad',
                                          '2n Cycle': 'Pre-Grad',
                                          'Basic': 'Pre-Grad'})


# For Comprhensibility
df = df.rename(columns={'MntWines': 'Wines',
                        'MntFruits': 'Fruits',
                        'MntMeatProducts': 'Meat',
                        'MntFishProducts': 'Fish',
                        'MntSweetProducts': 'Sweet',
                        'MntGoldProds': 'Gold'})

# Total number of people in a household
df['Children'] = df['Kidhome'] + df['Teenhome']
df['Total_Household'] = df['Relationship_Status'].replace({'Single': 1, 'Attached': 2}) + df['Children']

# Adding new column labeled 'Parent'
df['Parent'] = np.where(df['Children'] > 0, 1, 0)

# Num of days since last purchase (Recency)
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], format="%d-%m-%Y", dayfirst=True)
df['Recency'] = (pd.to_datetime('2024-01-01') - df['Dt_Customer']).dt.days

# Total number of purchases (Frequency)
df['Frequency'] = df[['NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases']].sum(axis=1)

# Total amount spent (Monetary)
df['Monetary'] = df[['Wines', 'Fruits', 'Meat', 'Fish', 'Sweet', 'Gold']].sum(axis=1)

# Average amount spent (Average Purchase)
df['Avg_Purchase_Value'] = df['Monetary'] / df['Frequency']

# Frequency of purchase (Loyalty)
df['Loyalty'] = np.where(df['Frequency'] > df['Frequency'].median(), 1, 0)

# Income per Total Household (Income per Total Household)
df['Income_per_Household_Member'] = df['Income'] / df['Total_Household']

# High Spender Indicator (High_Spender)
df['High_Spender'] = np.where(df['Monetary'] > df['Monetary'].median(), 1, 0)

# Indicate if a customer is responsive to promotions
df['Promo_Responsive'] = np.where(df['AcceptedCmp1'] + df['AcceptedCmp2'] + df['AcceptedCmp3'] + df['AcceptedCmp4'] + df['AcceptedCmp5'] > 0, 1, 0)

# Identify the preferred product category for each customer
df['Preferred_Category'] = df[['Wines', 'Fruits', 'Meat', 'Fish', 'Sweet', 'Gold']].idxmax(axis=1)

# Active Engagemnt Score
df['Engagement_Score'] = df['AcceptedCmp3'] + df['AcceptedCmp4'] + df['AcceptedCmp5'] + df['Response'] + df['Complain']
```

```python
# Dropping redundant feature.  (Do not drop anything for a full in-depth analysis)
df.drop(columns=['Kidhome', 'Teenhome', 'NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases', 
                 'AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'Income', 'Marital_Status', 
                 'Dt_Customer', 'Z_CostContact', 'Z_Revenue', 'Year_Birth', 'ID'], inplace=True)
```

# Visualising the Data

```python
# Check for infinite or NaN values and replace them
df.replace([np.inf, -np.inf], np.nan, inplace=True)
df.dropna(inplace=True)  # Remove rows with NaN values

# Labeling the numerical and categorical features
numerical_features = ['Age', 'Income_per_Household_Member', 'Frequency', 'Monetary', 'Avg_Purchase_Value']
categorical_features = ['Relationship_Status', 'Education', 'Parent', 'High_Spender', 'Promo_Responsive',
                        'Preferred_Category']

# Set up the plotting area
sns.set_style("darkgrid")

# Define a color palette
palette = sns.color_palette("husl", len(numerical_features))
```

## Histogram

```python
# Histograms for numerical features
plt.figure(figsize=(15, 10))
for i, feature in enumerate(numerical_features):
    plt.subplot(2, 3, i+1)
    sns.histplot(df[feature], color=palette[i], bins=30)
    plt.title(f'Histogram of {feature}')
plt.tight_layout()
plt.show()
```

# Pie Chart

```python
# Pie Chart for categorical features
# Set up the plotting area
plt.figure(figsize=(15, 15))

# Loop through each categorical feature to create pie charts
for i, feature in enumerate(categorical_features):
    plt.subplot(4, 2, i+1)
    # Compute value counts and plot pie chart
    counts = df[feature].value_counts()
    plt.pie(counts, labels=counts.index, autopct='%1.1f%%', colors=sns.color_palette("husl", len(counts)))
    plt.title(f'Pie Chart of {feature}')
    
plt.tight_layout()
plt.show()
```
